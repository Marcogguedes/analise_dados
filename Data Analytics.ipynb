## Tratamento e Mesclagem dos Datraframes

### Criando os dataframes

```python
# importando a biblioteca
import pandas as pd
```

```python
# carregando os datasets
vendas_df = pd.read_csv('Contoso - Vendas  - 2017.csv', sep=';')
produtos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep=';')
clientes_df = pd.read_csv('Contoso - Clientes.csv', sep=';')
lojas_df = pd.read_csv('Contoso - Lojas.csv', sep=';')
```

### Excluindo colunas

```python
# excluindo colunas do dataset clientes
clientes_df = clientes_df.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1)
```

```python
clientes_df.head()
```

### Filtrando colunas

```python
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]
```

```python
produtos_df.head()
```

### Juntando os dataframes para ter um único dataset com as variáveis necessárias.

```python
# utilizando o método (merge) para unir as variáveis e forma um único dataframe
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
```

```python
vendas_df.head()
```

```python
# renomeando o nome da coluna e-mail para e-mail do cliente utilizando um dicionário
vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail do Cliente'})
```

```python
vendas_df.head()
```

```python
vendas_df.info()
```

### Iniciando as análises

#### Qual cliente realizou o maior número de compras?

```python
frequencia_clientes = vendas_df['E-mail do Cliente'].value_counts()
```

```python
frequencia_clientes.head()
```

```python
# plotando as informações em gráfico
frequencia_clientes[:5].plot(figsize=(15,5))
```

#### Qual loja mais vendeu?

```python
vendas_lojas = vendas_df.groupby('Nome da Loja').sum()
```

```python
# filtrando as colunas
vendas_lojas = vendas_lojas[['Quantidade Vendida']]
```

```python
vendas_lojas
```

```python
# Ordenando o dataframe da loja com maior quantidade de venda para a menor loja
vendas_lojas = vendas_lojas.sort_values('Quantidade Vendida', ascending = False)
vendas_lojas.head()
```

```python
# plotando em um gráfico as cinco lojas com maiores vendas
vendas_lojas[:5].plot(figsize=(15,5), kind='bar')
```

```python
# imprimindo o maior valor e o índice da primeira loja
maior_valor = vendas_lojas['Quantidade Vendida'].max()
melhor_loja = vendas_lojas['Quantidade Vendida'].idxmax()
print(melhor_loja, maior_valor)
```

#### Qual o loja que menos vendeu?

```python
# imprimindo o menor valor e o índice da última loja
menor_valor = vendas_lojas['Quantidade Vendida'].min()
pior_loja = vendas_lojas['Quantidade Vendida'].idxmin()
print(pior_loja, menor_valor)
```
